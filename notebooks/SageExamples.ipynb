```python
# Import necessary SageMath libraries
from sage.all import *

# Define the variable
x = var('x')

# Define the function
f = sin(x) * exp(-x/2)

# Compute the derivative
f_derivative = diff(f, x)

# Compute the integral
f_integral = integrate(f, x)

# Plot the function, its derivative, and its integral
p1 = plot(f, (x, -2, 10), color='blue', legend_label='$f(x) = \sin(x) \cdot e^{-x/2}$')
p2 = plot(f_derivative, (x, -2, 10), color='green', legend_label='$f\'(x)$')
p3 = plot(f_integral, (x, -2, 10), color='red', legend_label='$\int f(x) dx$')

# Show the plots
show(p1 + p2 + p3, ymin=-2, ymax=2, legend_loc='upper right')

# Display the function, its derivative, and its integral in LaTeX
display(f)
display(f_derivative)
display(f_integral)
```
