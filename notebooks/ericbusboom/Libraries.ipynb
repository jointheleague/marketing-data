```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the main page with library locations
main_url = "https://www.sandiego.gov/public-library/locations"

# Send a request to fetch the page content
response = requests.get(main_url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the specific div containing the library links
libraries_section = soup.find('div', class_='clearfix text-formatted field field--name-body field--type-text-with-summary field--label-hidden field__item')
libraries = libraries_section.find_all('a', href=True)

# Extracting library names and URLs
library_data_parsed = []
for library in libraries:
    name = library.text.strip()
    url = f"https://www.sandiego.gov{library['href']}"
    library_data_parsed.append([name, url])

# Creating DataFrame
df_parsed = pd.DataFrame(library_data_parsed, columns=["Library Name", "URL"])

# Saving to CSV
csv_file_path_parsed = "Library_Locations_Parsed.csv"
df_parsed.to_csv(csv_file_path_parsed, index=False)

df_parsed
```
